```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import time
import quandl
import datetime as dt

authtoken= "qRy47RRsVxPSxUfSJgzy"
```

```python
def data_pull(quandl_code):
    df_name = pd.DataFrame(quandl.get(quandl_code,authtoken=authtoken)).reset_index()
    return df_name
```

```python
# University of Michigan Consumer Sentiment Index
df = data_pull("UMICH/SOC1").rename(columns={'Index':'Value'})
fig, ax = plt.subplots(1,1,figsize=(16,5))
sns.lineplot(data=df, x="Date", y="Value", color='r')
ax.set_title("US Consumer Sentiment",fontsize=14)
ax.set_xlim(['1-1-1950','12-31-2020'])
ax.set_ylim([40,120])
```
